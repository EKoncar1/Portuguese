```python
# Positive vs negative balance comparison

def balance_checker(amount):
    if amount < 0:
        return "Negative balance in account"
    elif amount > 0:
        return "Positive balance in account"
    else:
        return "Amount not found"
    
df["balance_checker"]=df['balance'].apply(balance_checker)
df["balance_checker"].value_counts().plot(kind="bar", legend=True)
plt.title("Distribution of Account Balances")
plt.xlabel("Account Status")
plt.ylabel("Status Count")
```

```python
# Negative balance in account per marital status group 
marital = df[df['balance_checker']=="Negative balance in account"].groupby("marital")
marital.agg({"marital":"count"}).plot(kind="bar", legend=True)
plt.title("Negative balance per Marital Status Group")
plt.xlabel("Marital Status")
plt.ylabel("Negative Balance Count")
```

```python
# Outcome variable count
df['y'].value_counts()
```

```python
# Function to change categorial to integer

def into_numbers(column_name):
    if column_name == "yes":
        return 1
    elif column_name=="no":
        return 0
    else:
        return "column name was not found"
```

```python
# Plot outcome count
df["Outcome_Variable"]= df['y'].apply(into_numbers)
df["Outcome_Variable"].value_counts().plot(kind='bar', legend=True)
plt.title("Outcome variable")
plt.xlabel("Outcome Variable")
plt.ylabel("Count of Outcome Variable")
```

```python
# Scatter plot of outcome variable
plt.scatter(x = df.age.astype("float"), y=df.Outcome_Variable.astype("float"))
plt.title("Scatter Plot of Outcome Variable")
plt.xlabel("Age")
plt.ylabel("Outcome Variable")
```

```python
# Train the model 
X_train, X_test, Y_train, Y_test = train_test_split(df[['age', 'balance']], df["Outcome_Variable"], test_size=0.1)
```

```python
# Run logistic regression and calculate accuracy of model 
model = LogisticRegression()
model.fit(X_train, Y_train)
model.predict(X_test)
model.score(X_test, Y_test)
```

```python
# Model probabilities
model.predict_proba(X_test)
```
